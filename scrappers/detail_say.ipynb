```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import time
from selenium.webdriver.firefox.options import Options as foptions
from selenium.webdriver.edge.options import Options as eoptions
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from lxml import etree

Foptions = foptions()
Foptions.headless = True

Eoptions = eoptions()
Eoptions.headless = True



unis = pd.read_excel('data/finished/unis_details_say_rj.xlsx')
try:
    unis = unis.drop(columns=['Unnamed: 0'])
except:
    pass


```

```python
for link in range(0,len(unis)):
        
        try:
            if len(unis['bolum_url'][link]) <5 or 'lightbox' in unis['bolum_url'][link]:
                unis['Yerleşen son kişinin Diploma notu'][link] = pd.NA
                unis['Ortalama OBP'][link] = pd.NA
                unis['TYT Temel Matematik'][link] = pd.NA
                unis['TYT Fen Bilimleri'][link] = pd.NA
                unis['TYT Türkçe'][link] = pd.NA
                unis['TYT Sosyal Bilimler'][link] = pd.NA
                unis['AYT Matematik'][link] = pd.NA
                unis['AYT Fizik'][link] = pd.NA
                unis['AYT Kimya'][link] = pd.NA
                unis['AYT Biyoloji'][link] = pd.NA
                unis['Profesör'][link] = pd.NA
                unis['Doçent'][link] = pd.NA
                unis['Doktora'][link] = pd.NA
                unis['Toplam Öğretim Görvelisi'][link] = pd.NA
                continue
            else:
                pass
        except:
            unis['Yerleşen son kişinin Diploma notu'][link] = pd.NA
            unis['Ortalama OBP'][link] = pd.NA
            unis['TYT Temel Matematik'][link] = pd.NA
            unis['TYT Fen Bilimleri'][link] = pd.NA
            unis['TYT Türkçe'][link] = pd.NA
            unis['TYT Sosyal Bilimler'][link] = pd.NA
            unis['AYT Matematik'][link] = pd.NA
            unis['AYT Fizik'][link] = pd.NA
            unis['AYT Kimya'][link] = pd.NA
            unis['AYT Biyoloji'][link] = pd.NA
            unis['Profesör'][link] = pd.NA
            unis['Doçent'][link] = pd.NA
            unis['Doktora'][link] = pd.NA
            unis['Toplam Öğretim Görvelisi'][link] = pd.NA
            continue
        if pd.notna(unis['Ortalama OBP'][link]) and unis['Ortalama OBP'][link] != "" or 'lightbox' in unis['bolum_url'][link]:
            print(link, ' geçildi')
            continue
        driver = webdriver.Edge(options=Eoptions)
        driver.get(unis['bolum_url'][link])
        driver.maximize_window()
        html = driver.page_source
        soup = BeautifulSoup(html,"html.parser")

        try:
            bypass = driver.find_element(By.XPATH,'/html/body/div[3]/div/span')
            bypass.click()
            time.sleep(2)
        except:
            pass        

        son_button = driver.find_element(By.XPATH,'/html/body/div[2]/div[1]/div[8]/div/div[1]/div[1]/a/h4')
        son_button.click()
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div[1]/div[8]/div/div[1]/div[2]/div/div/table/tbody/tr[4]/td[2]'))
        )
        son_button = driver.find_element(By.XPATH,'/html/body/div[2]/div[1]/div[8]/div/div[2]/div[1]/a/h4')
        son_button.click()
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div[1]/div[8]/div/div[2]/div[2]/div/div/table/tbody/tr[1]/td[2]'))
        )
        son_button = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[2]/div/div[1]/div[1]/a/h4')
        son_button.click()
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH,'//*[@id="icerik_2050"]/table/thead/tr/th[1]'))
        )
        time.sleep(2)

        try:
            
            son_puan_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[1]/div[8]/div/div[1]/div[2]/div/div/table/tbody/tr[4]/td[2]')).text
            unis['Yerleşen son kişinin Diploma notu'][link] = son_puan_var
        except:
            unis['Yerleşen son kişinin Diploma notu'][link] = pd.NA
        
        try:
            
            ort_obp_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[1]/div[8]/div/div[2]/div[2]/div/div/table/tbody/tr[1]/td[2]')).text
            tyt_mat_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[1]/div[8]/div/div[2]/div[2]/div/div/table/tbody/tr[3]/td[2]')).text
            tyt_fen_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[1]/div[8]/div/div[2]/div[2]/div/div/table/tbody/tr[4]/td[2]')).text
            tyt_turkce_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[1]/div[8]/div/div[2]/div[2]/div/div/table/tbody/tr[5]/td[2]')).text
            tyt_sosyal_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[1]/div[8]/div/div[2]/div[2]/div/div/table/tbody/tr[6]/td[2]')).text
            ayt_yabancı1_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[1]/div[8]/div/div[2]/div[2]/div/div/table/tbody/tr[7]/td[2]')).text
            ayt_yabancı2_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[1]/div[8]/div/div[2]/div[2]/div/div/table/tbody/tr[8]/td[2]')).text
            ayt_yabancı3_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[1]/div[8]/div/div[2]/div[2]/div/div/table/tbody/tr[9]/td[2]')).text
            ayt_yabancı4_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[1]/div[8]/div/div[2]/div[2]/div/div/table/tbody/tr[10]/td[2]')).text
            unis['Ortalama OBP'][link] = ort_obp_var
            unis['TYT Temel Matematik'][link] = tyt_mat_var
            unis['TYT Fen Bilimleri'][link] = tyt_fen_var
            unis['TYT Türkçe'][link] = tyt_turkce_var
            unis['TYT Sosyal Bilimler'][link] = tyt_sosyal_var
            unis['AYT Matematik'][link] = ayt_yabancı1_var
            unis['AYT Fizik'][link] = ayt_yabancı2_var
            unis['AYT Kimya'][link] = ayt_yabancı3_var
            unis['AYT Biyoloji'][link] = ayt_yabancı4_var
            
        except:
            unis['Ortalama OBP'][link] = pd.NA
            unis['TYT Temel Matematik'][link] = pd.NA
            unis['TYT Fen Bilimleri'][link] = pd.NA
            unis['TYT Türkçe'][link] = pd.NA
            unis['TYT Sosyal Bilimler'][link] = pd.NA
            unis['AYT Matematik'][link] = pd.NA
            unis['AYT Fizik'][link] = pd.NA
            unis['AYT Kimya'][link] = pd.NA
            unis['AYT Biyoloji'][link] = pd.NA
        
        try:
            
            prof_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[2]/div[2]/div/div[1]/div[2]/div/div/table/tbody/tr[1]/td[2]')).text
            doç_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[2]/div[2]/div/div[1]/div[2]/div/div/table/tbody/tr[2]/td[2]')).text
            doktora_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[2]/div[2]/div/div[1]/div[2]/div/div/table/tbody/tr[3]/td[2]')).text
            toplam_var = driver.find_element(By.XPATH,('/html/body/div[2]/div[2]/div[2]/div/div[1]/div[2]/div/div/table/tbody/tr[4]/td[2]')).text

            unis['Profesör'][link] = prof_var
            unis['Doçent'][link] = doç_var
            unis['Doktora'][link] = doktora_var
            unis['Toplam Öğretim Görvelisi'][link] = toplam_var
        except:
            pass
        driver.quit()

        if link % 5 ==0 :
            print(link, ' kadar tamamlandı')

        unis.to_excel('data/finished/unis_details_say_rj.xlsx',index=False)


```
